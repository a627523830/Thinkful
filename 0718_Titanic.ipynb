```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
df = pd.read_csv('titanic.csv')
print(df)
```

```python
any(df['Age'].isnull()) #Check if the age has missing values
```

```python
df.dropna(subset=['Age'], inplace=True) #Delete observations with missing age
```

```python
plt.scatter( x=df['Age'], y=df['Fare'],color='purple',marker='x', s=10) 
#We can see from the scatter plot that the number of people with fares below 100 dollars is far more than 
#the number of people with fares of more than 100 dollars.There is no significant relationship between fare and age。
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(20,5))
df["Age"].hist(ax=ax[0])
ax[0].set_title("Histogram plot of Age")
df["Fare"].hist(ax=ax[1])
ax[1].set_title("Histogram plot of Fare")
#It can be found that the majority of passengers are between the ages of 20 and 40 and are generally normally distributed. 
#The fare for most passengers is very low, between 0-100, and the fare for other passengers is higher.
```

```python
fig,ax = plt.subplots(figsize=(10,5))
df["Survived"].value_counts().plot(kind="bar")
ax.set_xticklabels(("Not Survived","Survived"),  rotation= "horizontal" )
ax.set_title("Bar plot for Survived ")
#It can be found that most of the passengers were not survived .
```

```python
fig,ax = plt.subplots(figsize=(10,5))
df["Sex"].value_counts().plot(kind="bar")
ax.set_xticklabels(("male","female"),rotation= "horizontal"  )
ax.set_title("Bar plot of Sex ")
#It can be found that most of the passengers are male.
```

```python
fig,ax = plt.subplots(figsize=(10,5))
df["Pclass"].value_counts().plot(kind="bar")
ax.set_xticklabels(("Third Class","First Clasee","Second Class"),rotation= "horizontal"  )
ax.set_title("Bar plot for Pclass ")
#It can be found that most of the passengers are in the third class, 
#and there are about 200 passengers in the first class and the second class.
```

```python
plt.boxplot(x = df['Age'], # Specify data
            patch_artist=True, # Requires a box to be filled with custom colors, default white fill
            showmeans=True, # Show the mean as a point
            boxprops = {'color':'black','facecolor':'#9999ff'}, # Set box properties, fill color and border color
            flierprops = {'marker':'o','markerfacecolor':'red','color':'black'}, # Set the outlier property, point shape, fill color, and border color
            meanprops = {'marker':'D','markerfacecolor':'indianred'}, # Set the properties of the mean point, the shape of the point, the fill color
            medianprops = {'linestyle':'--','color':'orange'}) # Set the properties of the median line, the type and color of the line
plt.ylim(0,85)
plt.show()
#For all passengers, it is easy to see from the picture that the average age of passengers is 30 years old,
#one quarter of passengers are less than 20 years old, and another quarter is more than 38 years old, 
#in other words, half People, ages between 20 and 38 years old; 
#from the mean (red diamonds) slightly above the median (yellow dotted line), indicating that age is biased and right off; 
#at the same time, we will Some red outliers were found, all of which were over 64 years old.
```

```python
# Store the age groups of different positions in the list Age variable through the for loop
Age = []
Levels = df.Pclass.unique()
for Pclass in Levels:
    Age.append(df.loc[df.Pclass==Pclass,'Age'])

plt.boxplot(x = Age, 
            patch_artist=True,
            labels = ['First Class','Second Class','Third Class'], 
            showmeans=True, 
            boxprops = {'color':'black','facecolor':'#9999ff'}, 
            flierprops = {'marker':'o','markerfacecolor':'red','color':'black'},
            meanprops = {'marker':'D','markerfacecolor':'indianred'},
            medianprops = {'linestyle':'--','color':'orange'})

plt.show()
#If we look at the different ages of the crowd, we will find a clear trend, that is, 
#the passengers with the higher class level, the higher their age, 
#the average age of the three classes is 38, 30 and 25, indicating that older the passengers are, the richer they will be, and the class they buy may be higher. 
#At the same time, in the second and third class, there are some abnormal users in the age of the passengers.
```
