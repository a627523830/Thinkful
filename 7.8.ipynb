```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
var_1 = np.random.normal(0 ,1 , 100)
print(var_1)
```

```python
df = pd.DataFrame({'rand': var_1})
df.hist()
plt.show()
```

```python
plt.hist(var_1, bins=20, color='c')
plt.axvline(var_1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(var_1.mean() + var_1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(var_1.mean()-var_1.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
var1 = np.random.normal(5,0.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+sd, color='green')
plt.axvline(x=mean-sd, color='green')
plt.show()
print('Results will vary depending on random seed. In general, stats should show some skew.')
```
